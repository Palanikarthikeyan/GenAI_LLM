```python
# Data Transformation
# ------------------------
# How to split by character-character Text Splitter
#  default delimiter \n\n - chunk length is measured by number of chars

## Step 1
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
```

```python
## Step 2
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
text=text_splitter.split_documents(docs)
text[0]
```

```python
## another way - normal filehandling - Text <or> Regular File 
with open("speech.txt") as fobj:
    s = fobj.read()
len(s)
```

```python
text_splitter = CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
text=text_splitter.create_documents([s])
print(text[0])
```

```python
# How to recursively split text by chars 

from langchain_text_splitters import RecursiveCharacterTextSplitter
text = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text = text.split_documents(docs)
text[0]
```
